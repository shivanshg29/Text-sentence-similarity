```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error, r2_score
from sentence_transformers import SentenceTransformer
```

```python
data = [
    ("A man is eating food.", "A man is eating a piece of bread.", 0.8),
    ("A woman is playing violin.", "A woman is playing the violin.", 1.0),
    ("A man is riding a horse.", "A man is driving a car.", 0.3),
    ("A child is playing in the park.", "A kid is playing outdoors.", 0.9),
    ("Two dogs are running.", "Two cats are sleeping.", 0.2),
]
```

```python
models = [
    "all-MiniLM-L6-v2",
    "all-mpnet-base-v2",
    "paraphrase-MiniLM-L3-v2",
    "distiluse-base-multilingual-cased-v2"
]
```

```python
results = []

for model_name in models:
    model = SentenceTransformer(model_name)
    start_time = time.time()

    predicted_similarities = []
    ground_truth = []

    for s1, s2, sim in data:
        emb1 = model.encode(s1, convert_to_tensor=True)
        emb2 = model.encode(s2, convert_to_tensor=True)
        score = cosine_similarity(emb1.cpu().reshape(1, -1), emb2.cpu().reshape(1, -1))[0][0]
        predicted_similarities.append(score)
        ground_truth.append(sim)

    elapsed_time = time.time() - start_time
    mse = mean_squared_error(ground_truth, predicted_similarities)
    r2 = r2_score(ground_truth, predicted_similarities)

    results.append([model_name, mse, r2, elapsed_time])
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create DataFrame
columns = ["Model", "MSE", "R2", "Time"]
df = pd.DataFrame(results, columns=columns)

# Criteria: [Lower MSE, Higher R2, Lower Time]
criteria = ["min", "max", "min"]
data_matrix = df.iloc[:, 1:].values

# Normalize using Min-Max Scaler
scaler = MinMaxScaler()
normalized_matrix = scaler.fit_transform(data_matrix)

# Weights (equal for simplicity)
weights = np.array([1/3, 1/3, 1/3])
weighted_matrix = normalized_matrix * weights

# Ideal Best and Worst
ideal_best = np.max(weighted_matrix, axis=0)
ideal_worst = np.min(weighted_matrix, axis=0)

# Distance to Ideal Best & Worst
d_plus = np.linalg.norm(weighted_matrix - ideal_best, axis=1)
d_minus = np.linalg.norm(weighted_matrix - ideal_worst, axis=1)

# Topsis Score
scores = d_minus / (d_plus + d_minus)
df['TOPSIS Score'] = scores
df['Rank'] = df['TOPSIS Score'].rank(ascending=False).astype(int)
df = df.sort_values("TOPSIS Score", ascending=False).reset_index(drop=True)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='TOPSIS Score', y='Model', data=df, palette='viridis')
plt.title("TOPSIS Ranking of Sentence Similarity Models")
plt.xlabel("Score")
plt.ylabel("Model")
plt.grid(True)
plt.tight_layout()
plt.show()
```
